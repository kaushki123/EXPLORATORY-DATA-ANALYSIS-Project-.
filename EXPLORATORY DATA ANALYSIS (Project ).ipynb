# 1) CHECK  &  FIX MISSING VALUES

```python
df.isnull().sum()
```

```python
# Calculate % of missing values per column
df.isnull().mean() * 100
```

```python
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
plt.title("Missing Value Heatmap")
plt.show()
```

 **Insight of heatmap**:-

*  heatmap quickly shows which columns have missing values and how widespread the missingness is.

# For example :

* Many yellow lines in the director column → That column has many missing values.
* Only a few yellow dots in rating or date_added → Those are mostly complete.



```python
# missing values in date_added column
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_added'].fillna(pd.Timestamp("2000-01-01"), inplace=True)  # E
```

```python
# missing values in country column
most_common_country = df['country'].mode()[0]
df['country'].fillna(most_common_country, inplace=True)
```

```python
# missing values in rating column
most_common_rating = df['rating'].mode()[0]
df['rating'].fillna(most_common_rating, inplace=True)
```

```python
# missing values in duration column
df['duration'] = df.groupby('type')['duration'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown')
)
```


# 2) DATE SETTING


```python
df['month_added'] = df['date_added'].dt.month
```

```python
df['Day'] = df['date_added'].dt.day
```

```python
df['year_added'] = df['date_added'].dt.year
```

#3) EXTRACTING THE FIRST GENRE FROM THE LISTED_IN



```python
df['genre'] = df['listed_in'].str.split(',').str[0]  # first genre only
```

4) RATING CONVERT INTO RATING_GRADE

```python
rating_grades = {
    'TV-Y': 'A+',
    'TV-Y7': 'A',
    'TV-Y7-FV': 'A-',
    'G': 'B+',
    'TV-G': 'B+',
    'PG': 'B',
    'TV-PG': 'B-',
    'PG-13': 'C+',
    'TV-14': 'C',
    'R': 'D',
    'TV-MA': 'D-',
    'NC-17': 'F',
    'NR': 'N/A',
    'UR': 'N/A'
}

df['rating_grade'] = df['rating'].map(rating_grades)
```

```python
df.head()
```

## 5) DROP

```python
df.drop(columns=['director', 'cast'] , inplace=True)
```

# DESCRIPTION CONVERT INTO RATING_DESCRIPTION


```python
rating_descriptions = {
    'TV-Y': 'Suitable for all children',
    'TV-Y7': 'Suitable for older children',
    'TV-Y7-FV': 'Older children (fantasy violence)',
    'G': 'General audience',
    'TV-G': 'General audience',
    'PG': 'Parental guidance suggested',
    'TV-PG': 'Parental guidance suggested',
    'PG-13': 'Teens, parental caution',
    'TV-14': 'May not be suitable under 14',
    'R': 'Restricted, 17+',
    'TV-MA': 'Adults only (mature content)',
    'NC-17': 'No one 17 and under admitted',
    'NR': 'Not rated',
    'UR': 'Unrated'
}

df['rating_description'] = df['rating'].map(rating_descriptions)

#  this using for fix missing values:
df['rating_grade'].fillna('N/A', inplace=True)
df['rating_description'].fillna('Not rated / Unknown content suitability', inplace=True)
```

```python
type_counts = df['type'].value_counts()

# Use Seaborn style and palette
sns.set_style("whitegrid")
colors = sns.color_palette('pastel')

plt.figure(figsize=(7,7))
plt.pie(
    type_counts.values,
    labels=type_counts.index,
    autopct='%.1f%%',
    colors=colors,
    startangle=90,
    explode=[0.1 if i==type_counts.idxmax() else 0 for i in type_counts.index]
)
plt.title('Netflix: Movies vs TV Shows')
plt.axis('equal')  # Makes it a circle
plt.show()
```

18. Pie Chart: Movies vs TV Shows

* This highlights the proportion of each content type—commonly around a 70/30 split .
* Movies is **69.6%** & TV Shows is **30.4%** .

```python
# Bar Chart: Content Added by Year & Type
df['year_added'] = pd.to_datetime(df['date_added']).dt.year
counts = df.groupby(['year_added','type']).size().reset_index(name='count')

fig, ax = plt.subplots(figsize=(10,6))
for category, color in zip(['Movie','TV Show'], ['#E50914', '#221F1F']):
    sub = counts[counts['type'] == category]
    ax.plot(sub['year_added'], sub['count'], label=category, color=color, marker='o')

ax.set_xlabel('Year Added')
ax.set_ylabel('Number of Titles')
ax.legend()
plt.title('Netflix Content Type by Year Added')
plt.show()
```

19. This line plot reflects Netflix's shift toward TV Shows in recent years:

*  **Insight** 1 :-Since 2016–2017, the TV Show curve rises more steeply than the Movie line—

>  indicating a faster expansion in series content. While movies remained dominant in raw volume, TV Shows grew at a significantly faster rate, suggesting a strategic shift into episodic content

* **Insight** 2:-By around 2017, the share of new TV Shows being added increased notably. In earlier years, movies comprised ~58% of additions, but after 2017, they rose to ~70% of new content additions . This shows Netflix is not just adding more content overall

>  it’s focusing increasingly on original series.



```python
sns.set_style("whitegrid")
palette = sns.color_palette('pastel')

# Calculate the top listed genres for each content type
top_counts = df.groupby('type')['listed_in'].apply(
    lambda x: x.str.split(', ').explode().value_counts().head(5)
).reset_index(name='count')

fig, axes = plt.subplots(1, 2, figsize=(14, 7))

for ax, content_type in zip(axes, ['Movie', 'TV Show']):
    subset = top_counts[top_counts['type'] == content_type]
    ax.pie(
        subset['count'],
        labels=subset['level_1'],
        autopct='%.1f%%',
        colors=palette,
        startangle=90,
        explode=[0.1] + [0]* (len(subset)-1)
    )
    ax.set_title(f"Top Genres in {content_type}s")
    ax.axis('equal')

plt.show()
```

3) Multivariate Analysis

```python
# Facet Grid

grouped = df.groupby(['country', 'type', 'rating_grade']).size().reset_index(name='count')

# Filter top countries
top_countries = df['country'].value_counts().head(6).index
filtered = grouped[grouped['country'].isin(top_countries)]

g = sns.catplot(
    data=filtered,
    x='rating_grade', y='count',
    col='country', hue='type',
    kind='bar', height=4, aspect=0.9,
    palette='Set2'
)

g.set_titles('{col_name}')
g.set_axis_labels('Rating Grade', 'Number of Titles')
g.fig.subplots_adjust(top=0.85)
g.fig.suptitle('Rating Grade Distribution by Country and Content Type')

plt.show()
```
